```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy import ndimage
```

```python
comet = cv2.imread('comet.jpg')
comet = cv2.cvtColor(comet,cv2.COLOR_RGB2BGR)
plt.imshow(comet)
print(comet.shape)
```

```python
comet_gray = np.copy(comet)
comet_gray = cv2.cvtColor(comet_gray,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(comet_gray,cmap='gray')
```

```python
ndimage.measurements.center_of_mass(comet_gray)
```

```python
com = ndimage.measurements.center_of_mass(comet_gray)
comx = int(ndimage.measurements.center_of_mass(comet_gray)[1])
comy = int(ndimage.measurements.center_of_mass(comet_gray)[0])
```

```python
color = (255,0,0)
t = 10

e = 150
n1 = (comx-e,comy-e) 
n2 = (comx+e,comy+e)
```

```python
comet_2 = np.copy(comet)
comet_2 = cv2.rectangle(comet_2,n1,n2,color,t)
```
